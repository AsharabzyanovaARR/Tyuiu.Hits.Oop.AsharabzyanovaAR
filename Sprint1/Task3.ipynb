#### Дополнительное задание
Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public class Student
{
    public string Name { get; protected set; }
    public int Age { get; protected set; }
    public int Course { get; protected set; }

    public Student(string name, int age, int course)
    {
        Name = name;
        Age = age;
        Course = course;
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится.");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен.");
    }

    public virtual void StudentInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age}, Курс: {Course}");
    }
}

public class BachelorStudent : Student
{
    public string Specialty { get; private set; }

    public BachelorStudent(string name, int age, int course, string specialty)
        : base(name, age, course)
    {
        Specialty = specialty;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} (бакалавр) изучает предметы по специальности '{Specialty}'.");
    }

    public override void StudentInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age}, Курс: {Course}, Специальность: {Specialty}");
    }
}

public class MasterStudent : Student
{
    public string ScientificAdvisor { get; private set; }

    public MasterStudent(string name, int age, int course, string scientificAdvisor)
        : base(name, age, course)
    {
        ScientificAdvisor = scientificAdvisor;
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} (магистрант) сдаёт экзамен под руководством {ScientificAdvisor}.");
    }

    public override void StudentInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age}, Курс: {Course}, Научный руководитель: {ScientificAdvisor}");
    }
}

public class PhDStudent : Student
{
    public string DissertationTopic { get; private set; }

    public PhDStudent(string name, int age, int course, string dissertationTopic)
        : base(name, age, course)
    {
        DissertationTopic = dissertationTopic;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} (аспирант) проводит исследование по теме '{DissertationTopic}'.");
    }

    public override void StudentInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age}, Курс: {Course}, Тема диссертации: '{DissertationTopic}'");
    }
}

public class University
{
    private List<Student> students = new List<Student>();
    
    public void AddStudent(Student student)
    {
        students.Add(student);
        Console.WriteLine($"Добавлен студент: {student.Name}");
    }
    
    public void OrganizeConsultation(string bachelorName, string masterName, string topic)
    {
        var bachelor = students.Find(s => s.Name == bachelorName && s is BachelorStudent);
        var master = students.Find(s => s.Name == masterName && s is MasterStudent);
        
        if (bachelor != null && master != null)
        {
            Console.WriteLine($"\nКонсультация: {bachelorName} (бакалавр) консультируется у {masterName} (магистрант)");
            Console.WriteLine($"Тема: {topic}");
            
            bachelor.Study();
            ((MasterStudent)master).TakeExam();
        }
        else
        {
            Console.WriteLine("Ошибка: студенты не найдены.");
        }
    }
    
    public void ShowAllStudents()
    {
        Console.WriteLine("\nВсе студенты:");
        foreach (var student in students)
        {
            student.StudentInfo();
        }
    }
}

var university = new University();

var student1 = new Student("Иван Иванов", 19, 2);
var student2 = new BachelorStudent("Анна Смирнова", 20, 3, "Информатика");
var student3 = new MasterStudent("Петр Петров", 23, 1, "Проф. Сидоров");
var student4 = new PhDStudent("Мария Козлова", 26, 4, "Искусственный интеллект в образовании");

university.AddStudent(student1);
university.AddStudent(student2);
university.AddStudent(student3);
university.AddStudent(student4);

university.OrganizeConsultation("Анна Смирнова", "Петр Петров", "Базы данных");

university.ShowAllStudents();

Console.WriteLine("\nОперации со студентами:");
student1.Study();
student2.TakeExam();
student3.Study();
student4.TakeExam();
```
