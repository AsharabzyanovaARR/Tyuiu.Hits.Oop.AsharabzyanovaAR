```
// Простое наследование: Student -> Person
public class Person
{
    public string Name { get; protected set; }
    public int Age { get; protected set; }
    
    public Person(string name, int age)
    {
        Name = name;
        Age = age;
    }
    
    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Человек: {Name}, Возраст: {Age}");
    }
}

// Простое наследование: Student наследует от Person
public class Student : Person
{
    public int Course { get; protected set; }
    
    public Student(string name, int age, int course) : base(name, age)
    {
        Course = course;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Студент: {Name}, Возраст: {Age}, Курс: {Course}");
    }
    
    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится на {Course} курсе.");
    }
    
    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен на {Course} курсе.");
    }
}

// Сложное наследование: BachelorStudent -> Student -> Person
public class BachelorStudent : Student
{
    public string Specialty { get; private set; }
    
    public BachelorStudent(string name, int age, int course, string specialty) 
        : base(name, age, course)
    {
        Specialty = specialty;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Бакалавр: {Name}, Возраст: {Age}, Курс: {Course}, Специальность: {Specialty}");
    }
    
    public override void Study()
    {
        Console.WriteLine($"{Name} (бакалавр) изучает предметы по специальности '{Specialty}'.");
    }
}

// Сложное наследование: MasterStudent -> Student -> Person
public class MasterStudent : Student
{
    public string ScientificAdvisor { get; private set; }
    
    public MasterStudent(string name, int age, int course, string scientificAdvisor) 
        : base(name, age, course)
    {
        ScientificAdvisor = scientificAdvisor;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Магистрант: {Name}, Возраст: {Age}, Курс: {Course}, Научный руководитель: {ScientificAdvisor}");
    }
    
    public override void TakeExam()
    {
        Console.WriteLine($"{Name} (магистрант) сдаёт экзамен под руководством {ScientificAdvisor}.");
    }
}

// Сложное наследование: PhDStudent -> Student -> Person
public class PhDStudent : Student
{
    public string DissertationTopic { get; private set; }
    
    public PhDStudent(string name, int age, int course, string dissertationTopic) 
        : base(name, age, course)
    {
        DissertationTopic = dissertationTopic;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Аспирант: {Name}, Возраст: {Age}, Курс: {Course}, Тема диссертации: '{DissertationTopic}'");
    }
    
    public override void Study()
    {
        Console.WriteLine($"{Name} (аспирант) проводит исследование по теме '{DissertationTopic}'.");
    }
}

// Интерфейс 1 для множественного наследования
public interface IResearcher
{
    void ConductResearch();
    string ResearchTopic { get; }
}

// Интерфейс 2 для множественного наследования
public interface ITeacherAssistant
{
    void ConductSeminar(string subject);
    void GradeAssignments(int count);
}

// Интерфейс 3 для множественного наследования
public interface IScholarshipHolder
{
    decimal ScholarshipAmount { get; }
    void ApplyForScholarship();
    void ReceiveScholarship();
}

// Класс для множественного наследования через интерфейсы
public class GraduateStudent : Student, IResearcher, ITeacherAssistant, IScholarshipHolder
{
    public string ResearchTopic { get; private set; }
    public decimal ScholarshipAmount { get; private set; }
    
    public GraduateStudent(string name, int age, int course, string researchTopic) 
        : base(name, age, course)
    {
        ResearchTopic = researchTopic;
        ScholarshipAmount = 0;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Выпускник-исследователь: {Name}, Возраст: {Age}, Курс: {Course}, Тема исследования: '{ResearchTopic}', Стипендия: {ScholarshipAmount:C}");
    }
    
    // Реализация методов из IResearcher
    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит исследование по теме: {ResearchTopic}");
    }
    
    // Реализация методов из ITeacherAssistant
    public void ConductSeminar(string subject)
    {
        Console.WriteLine($"{Name} проводит семинар по предмету: {subject}");
    }
    
    public void GradeAssignments(int count)
    {
        Console.WriteLine($"{Name} проверяет задания у {count} студентов");
    }
    
    // Реализация методов из IScholarshipHolder
    public void ApplyForScholarship()
    {
        Console.WriteLine($"{Name} подает заявку на стипендию");
    }
    
    public void ReceiveScholarship()
    {
        ScholarshipAmount = 15000;
        Console.WriteLine($"{Name} получил стипендию в размере {ScholarshipAmount:C}");
    }
}

// Класс для демонстрации простого наследования (без Student)
public class Professor : Person
{
    public string Department { get; private set; }
    
    public Professor(string name, int age, string department) : base(name, age)
    {
        Department = department;
    }
    
    public override void DisplayInfo()
    {
        Console.WriteLine($"Профессор: {Name}, Возраст: {Age}, Кафедра: {Department}");
    }
    
    public void Teach(string subject)
    {
        Console.WriteLine($"{Name} преподает предмет: {subject}");
    }
}

public class University
{
    private List<Person> people = new List<Person>();
    
    public void AddPerson(Person person)
    {
        people.Add(person);
        Console.WriteLine($"Зачислен/Принят: {person.GetType().Name} - {person.Name}");
    }
    
    public void ShowAll()
    {
        Console.WriteLine("\nВсе в университете:");
        foreach (var person in people)
        {
            person.DisplayInfo();
            
            if (person is GraduateStudent graduate)
            {
                graduate.ConductResearch();
                graduate.ConductSeminar("Математика");
                graduate.ReceiveScholarship();
            }
            
            if (person is Student student)
            {
                student.Study();
                student.TakeExam();
            }
            
            if (person is BachelorStudent bachelor)
            {
                Console.WriteLine($"Бакалавр специализируется на: {bachelor.Specialty}");
            }
            else if (person is MasterStudent master)
            {
                Console.WriteLine($"Магистрант работает с руководителем: {master.ScientificAdvisor}");
            }
            else if (person is PhDStudent phd)
            {
                Console.WriteLine($"Аспирант исследует: {phd.DissertationTopic}");
            }
            else if (person is Professor professor)
            {
                professor.Teach("Программирование");
            }
            
            Console.WriteLine();
        }
    }
}

University university = new University();

// Простое наследование (Professor -> Person)
university.AddPerson(new Professor("Александр Иванов", 55, "Кафедра информатики"));

// Простое наследование (Student -> Person)
university.AddPerson(new Student("Иван Иванов", 19, 2));

// Сложное наследование (BachelorStudent -> Student -> Person)
university.AddPerson(new BachelorStudent("Анна Смирнова", 20, 3, "Информатика"));

// Сложное наследование (MasterStudent -> Student -> Person)
university.AddPerson(new MasterStudent("Петр Петров", 23, 1, "Проф. Сидоров"));

// Сложное наследование (PhDStudent -> Student -> Person)
university.AddPerson(new PhDStudent("Мария Козлова", 26, 4, "Искусственный интеллект"));

// Множественное наследование (через интерфейсы)
university.AddPerson(new GraduateStudent("Дмитрий Соколов", 24, 2, "Квантовые вычисления"));

university.ShowAll();

Console.WriteLine("\n=== Демонстрация полиморфизма ===");
Person[] persons = new Person[]
{
    new Student("Алексей Волков", 21, 3),
    new BachelorStudent("Екатерина Морозова", 22, 4, "Экономика"),
    new MasterStudent("Артем Лебедев", 25, 2, "Проф. Козлов"),
    new PhDStudent("Наталья Васнецова", 28, 5, "Биоинформатика"),
    new GraduateStudent("Павел Орлов", 27, 3, "Робототехника"),
    new Professor("Виктор Павлов", 60, "Кафедра математики")
};

foreach (var person in persons)
{
    Console.WriteLine($"\n{person.GetType().Name}:");
    person.DisplayInfo();
}
```
